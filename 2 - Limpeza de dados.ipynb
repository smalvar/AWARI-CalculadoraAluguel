```python
import pandas as pd
```

```python
df = pd.read_csv('banco_alugueis.csv',dtype = str)
df = df.drop(['Unnamed: 0'],axis=1)
```

```python
df.head(10)
```

```python
df.columns
```

```python
all_columns = list(df)
df[all_columns] = df[all_columns].astype(str)
```

```python
df.dtypes
```

```python
df['area'] = df['area'].str.extract('(.*) m²')
```

```python
df.head()
```

```python
df['vaga'] = df['vaga'].str.extract('(.*) Vaga')
```

```python
df.head()
```

```python
df['banheiro'] = df['banheiro'].str.extract('(.*) Banheiro')
df['quartos'] = df['quartos'].str.extract('(.*) Quarto')
df.head()
```

```python
df.shape
```

```python
df['aluguel'] = df['aluguel'].str.extract('R\$ (.*) /Mês')
df['aluguel'] = df['aluguel'].replace('\.', '', regex=True)
df.head()
```

```python
df['bairro'].unique()
```

```python
df.info()
```

```python
df.describe()
```

```python
cols = df.columns.drop(['endereco','bairro','rua'])
```

```python
cols
```

```python
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
```

```python
df.info()
```
